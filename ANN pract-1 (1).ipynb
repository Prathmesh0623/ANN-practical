1. Write a Python program to plot a few activation functions that are being used in neural networks.

```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1.Binery step

```python
def binaryStep(x):
    return np.heaviside(x,1)
```

```python
x=np.linspace(-10,10)
plt.plot(x,binaryStep(x))
plt.axis('tight')
plt.title('Activation function:bineryStep')
plt.grid(True)
plt.show()
```

# 2.Sigmoid

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x=np.linspace(-10,10)
plt.plot(x,sigmoid(x))
plt.axis('tight')
plt.title('Activation function : Sigmoid')
plt.grid(True)
plt.show
```

# 3.tanh(Hyperbolic)

```python
def tanh_fun(x):
    return np.tanh(x)
```

```python
x=np.linspace(-10,10)
plt.plot(x,tanh_fun(x))
plt.axis('tight')
plt.title('Activation function : tanh(hyperbolic)')
plt.grid(True)
```

# 4.Softmax

```python
def Softmax(x):
    return np.exp(x)/np.sum(np.exp(x),axis=0)
```

```python
x=np.linspace(-5,5)
plt.plot(x,Softmax(x))
plt.axis('tight')
plt.title('Activation function : Softmax')
plt.grid(True)
plt.show()
```

# 5.Exponential Linear Unit

```python
def expnlu(x):
    return 1.0*(np.exp(x)-1)
```

```python
x=np.linspace(-10,10)
plt.plot(x,expnlu(x))
plt.axis('tight')
plt.title('Activation function :Exponential Linear Unit')
plt.grid(True)
plt.show()
```

# 6.RelU

```python
def relu(x):
    x1=[]
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)

    return x1
   
```

```python
x=np.linspace(-10,10)
plt.plot(x,relu(x))
plt.axis('tight')
plt.title('Activation function :Relu')
plt.grid(True)
plt.show()
```

# 7.Leaky Rule Function

```python
def lrf(x):
    x1=[]
    for i in x:
        if i>=0:
            x1.append(i)
        else:
            x1.append(0.01*i)
            
    return x1
```

```python
x=np.linspace(-5,5)
plt.plot(x,lrf(x))
plt.axis('tight')
plt.title('Activation Function :Leaky Rule Function ')
plt.show()
```

# 8.Rectified Linear

```python
def rectf(x):
    x1=[]
    for i in x:
        if i>=0:
            x1.append(i)
        else:
            x1.append(0)
    return x1
            
```

```python
x=np.linspace(-5,5)
plt.plot(x,rectf(x))
plt.axis('tight')
plt.title('Activation Function :Rectified Linear ')
plt.show()
```
