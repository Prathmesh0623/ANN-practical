# perceptron learning law
4. With a suitable example demonstrate the perceptron learning law with its decision regions using
python. Give the output in graphical form.


# note. use only one method to show region in exam

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
array=[]
threshold=1
def and_not(x1,x2):
    weights=[-1,3]
    inputs=np.array([x1,x2])
    product=np.dot(weights,inputs)
    array.append(product)
and_not(1,1)
and_not(1,0)
and_not(0,1)
and_not(0,0)
print(array)
```

# 1st method to show region

```python
plt.plot(array)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title(' perceptron learning law with its decision regions')
plt.grid(True)
plt.gca().spines['left'].set_position(('data', 0))
plt.gca().spines['bottom'].set_position(('data', 0))
plt.gca().spines['right'].set_color('none')
plt.gca().spines['top'].set_color('none')
plt.show()
```

# 2nd method

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_xlim([-6, 6])
ax.set_ylim([-6, 6])
ax.scatter(array[0], array[0], color='r', label=str(array[0]))
ax.scatter(-array[1], array[1], color='g', label=str(array[1]))
ax.scatter(-array[2], -array[2], color='b', label=str(array[2]))
ax.scatter(array[3], -array[3], color='y', label=str(array[3]))
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title(' perceptron learning law with its decision regions')
ax.legend()
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```
